```python
locations['site_number'] = locations['site_number'].astype(str).str.zfill(2)
accounts['site_number'] = accounts['site_number'].astype(str).str.zfill(2)
```

```python
accounts['customer_no'] = accounts['customer_no'].astype(str).str.zfill(7)
accounts['customer_key']=  accounts['site_number'] + "-" + accounts['customer_no']
```

```python
accounts.drop(['customer_no'], axis=1, inplace=True)
accounts.rename(columns={'customer_key': 'customer_no'}, inplace=True)
```

```python
# Calculate Customer Engagement Score
def calculate_engagement_score(accounts_df):
    accounts_df['last_engagement_date'] = pd.to_datetime(accounts_df['last_engagement_date'])
    accounts_df['recent_purchase_date'] = pd.to_datetime(accounts_df['recent_purchase_date'])
    
    # Calculate recency (days since last engagement and purchase)
    accounts_df['days_since_engagement'] = (pd.to_datetime('today') - accounts_df['last_engagement_date']).dt.days
    accounts_df['days_since_purchase'] = (pd.to_datetime('today') - accounts_df['recent_purchase_date']).dt.days

    # Engagement score formula (adjust weightings as needed)
    accounts_df['engagement_score'] = (0.5 / (accounts_df['days_since_engagement'] + 1)) + (0.5 / (accounts_df['days_since_purchase'] + 1))
    
    return accounts_df[['customer_no', 'engagement_score']]

engagement_scores = calculate_engagement_score(accounts)
metric_engagement_score=engagement_scores['engagement_score'].mean()
```

```python
# Calculate Customer Lifetime Value (CLV)
def calculate_clv(sales_df):
    # Calculate total revenue per customer
    sales_df['total_revenue'] = sales_df["ext_net"]
    
    # Aggregate total revenue per customer
    clv = sales_df.groupby('customer_no')['total_revenue'].sum().reset_index()
    clv.rename(columns={'total_revenue': 'customer_lifetime_value'}, inplace=True)
    
    return clv

clv = calculate_clv(sales)
metric_clv=clv['customer_lifetime_value'].mean()
metric_clv
```

```python
# Calculate Product Affinity Score
def calculate_product_affinity(sales_df, products_df):
    merged_df = pd.merge(sales_df, products_df, on='item_no')
    
    # Calculate total spend per product category
    merged_df['total_spend'] = merged_df['ext_net']
    
    affinity = merged_df.groupby(['customer_no', 'prod_category'])['total_spend'].sum().reset_index()
    affinity.rename(columns={'total_spend': 'product_affinity_score'}, inplace=True)
    
    return affinity

product_affinity = calculate_product_affinity(sales, products)
metric_product_affinity=product_affinity['product_affinity_score'].mean()
```

```python
# Calculate Recent Purchase Recency
def calculate_recent_purchase_recency(sales_df, accounts_df):
    sales_df['purchase_date'] = pd.to_datetime(sales_df['invoice_dt_sk'])
    
    # Get the latest purchase date per customer
    recent_purchase = sales_df.groupby('customer_no')['purchase_date'].max().reset_index()
    
    # Merge with accounts to get customer details
    recent_purchase = pd.merge(recent_purchase, accounts_df[['customer_no', 'customer_name']], on='customer_no')
    
    # Calculate recency (days since last purchase)
    recent_purchase['days_since_last_purchase'] = (pd.to_datetime('today') - recent_purchase['purchase_date']).dt.days
    
    return recent_purchase[['customer_no', 'days_since_last_purchase']]

recent_purchase_recency = calculate_recent_purchase_recency(sales, accounts)
metric_recent_purchase_recency=recent_purchase_recency['days_since_last_purchase'].mean()
```

```python
# Calculate Average Order Value (AOV)
def calculate_aov(sales_df):
    # Calculate total order value per customer
    sales_df['order_value'] = sales_df['ext_net']
    # Group by customer and calculate average order value
    aov = sales_df.groupby('customer_no')['order_value'].mean().reset_index()
    aov.rename(columns={'order_value': 'average_order_value'}, inplace=True)
    
    return aov

aov = calculate_aov(sales)
metric_aov=aov['average_order_value'].mean()
```

```python
# Calculate Repeat Purchase Rate
def calculate_repeat_purchase_rate(sales_df):
    repeat_purchase = sales_df.groupby('customer_no')['invoice_no'].nunique().reset_index()
    
    # Count repeat buyers (those with more than 1 unique invoice)
    repeat_purchase['repeat_purchase'] = repeat_purchase['invoice_no'].apply(lambda x: 1 if x > 1 else 0)
    
    repeat_rate = repeat_purchase['repeat_purchase'].mean()
    return repeat_rate

repeat_purchase_rate = calculate_repeat_purchase_rate(sales)
```

```python
# Calculate Conversion Rate
campaign_conversion['conversion_rate'] = (campaign_conversion['total_conversions'] / campaign_conversion['total_customers_targeted']) * 100
metric_campaign_conversion=campaign_conversion['conversion_rate'].mean()
```

```python
# Calculate Customer Retention Rate
def calculate_customer_retention_rate(sales):
    # Group sales by customer_no to count the number of purchases per customer
    customer_purchase_count = sales.groupby('customer_no').size()

    # Retained customers are those who have made more than one purchase
    retained_customers = customer_purchase_count[customer_purchase_count > 1]

    # Calculate number of retained customers and total customers
    retained_customers_count = len(retained_customers)
    total_customers_count = len(sales['customer_no'].unique())

    # Calculate customer retention rate
    retention_rate = (retained_customers_count / total_customers_count) * 100
    return retention_rate
metric_customer_retention_rate = calculate_customer_retention_rate(sales)
```

```python
# Calculate engagement channel effectiveness
def calculate_engagement_channel_effectiveness(sales, accounts):
    # Merge the accounts data with sales data to get the engagement channel for each customer in sales
    sales_with_channel = sales.merge(accounts[['customer_no', 'engagement_channel']], on='customer_no')

    # Filter out successful conversions (assuming ext_net > 0 indicates a successful sale)
    successful_sales = sales_with_channel[sales_with_channel['ext_net'] > 0]

    # Count the number of unique customers that made a purchase for each engagement channel
    conversions_by_channel = successful_sales.groupby('engagement_channel')['customer_no'].nunique()

    # Count the total number of unique customers engaged by each channel
    total_engagements_by_channel = accounts.groupby('engagement_channel')['customer_no'].nunique()

    # Calculate the engagement channel effectiveness
    channel_effectiveness = (conversions_by_channel / total_engagements_by_channel) * 100
    return channel_effectiveness

engagement_channel_effectiveness = calculate_engagement_channel_effectiveness(sales, accounts)
metric_engagement_channel_effectiveness=engagement_channel_effectiveness.mean()
```

```python
# Calculate Customer Churn Rate
def calculate_churn_rate(sales_df):
    customer_purchase_count = sales_df.groupby('customer_no').size()
    churned_customers = customer_purchase_count[customer_purchase_count == 1]
    churned_customers_count = len(churned_customers)
    total_customers_count = len(sales_df['customer_no'].unique())
    churn_rate = (churned_customers_count / total_customers_count) * 100
    return churn_rate
metric_churn_rate = calculate_churn_rate(sales)
```

```python
# Calculate Average Revenue Per User (ARPU)
def calculate_arpu(sales_df):
    total_revenue = sales_df['ext_net'].sum()
    total_customers = sales_df['customer_no'].nunique()
    arpu = total_revenue / total_customers
    return arpu
metric_arpu = calculate_arpu(sales)
```
