```python
import pandas as pd
from datetime import datetime, timedelta

# Load the CSV file
file_path = '/home/shalavyaagrawal/Documents/Prometheus/anamoly and dimension attribution/monthly_data.csv'
df = pd.read_csv(file_path)

# Function to generate dates for a given month
def generate_dates(year, month):
    start_date = datetime(year, month, 1)
    end_date = (start_date + timedelta(days=32)).replace(day=1) - timedelta(days=1)
    return [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

# Update the dates in the DataFrame
def update_dates(df, year, month):
    dates = generate_dates(year, month)
    date_index = 0
    for i in range(len(df)):
        df.at[i, 'Month'] = dates[date_index].strftime('%Y-%m-%d')
        date_index = (date_index + 1) % len(dates)

# Update dates for November 2024
update_dates(df, 2024, 11)

# Save the updated DataFrame back to the CSV file
df.to_csv(file_path, index=False)
```
